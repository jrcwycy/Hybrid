# Load resources

```python
def load_pathway(fpath):
    result = []
    with open(fpath) as f:
        for line in f:
            split_line = [x for x in line.strip().split('\t') if x]  # Remove empty strings directly

            row = {'label': split_line[0]}
            for gene in split_line[1:]:
                row[gene] = 1

            result.append(row)

    df = pd.DataFrame(result)
    df = df.fillna(0.0).set_index('label').astype(bool).T  # Chained operations for clarity

    return df

mpath = "PanglaoDB_Augmented_2021.txt"
pang = load_pathway(mpath)

cell_types = [
    'Fibroblasts',
    'Myocytes',
    'Myoblasts',
    'Myofibroblasts',
]

subset = pang[cell_types]
panglao_markers = subset[subset.any(axis=1)]

panglao_markers = panglao_markers.reset_index(names='gene')

print(panglao_markers.shape)
panglao_markers.head()

# fib_genes = panglao_markers['Fibroblasts'] == True
# fib_genes =  panglao_markers[fib_genes].index
# print(f"N fib genes: {len(fib_genes)}")

# myo_genes_panglao = (panglao_markers['Myocytes'] == True) | (panglao_markers['Myoblasts'] == True)
# myo_genes_panglao =  panglao_markers[myo_genes_panglao].index
# print(f"N myocyte/myoblast genes: {len(myo_genes_panglao)}")
```

# Combine 

```python
def markers_wide_to_long(df, gene_col="gene"):
    cell_cols = df.columns.drop(gene_col)
    long = (df
            .melt(id_vars=gene_col, value_vars=cell_cols,
                  var_name="cell_type", value_name="is_marker"))
    return long[long["is_marker"]].drop(columns="is_marker").reset_index(drop=True)


long1 = markers_wide_to_long(panglao_markers).assign(source='Panglao')
long2 = markers_wide_to_long(cellmarker_markers).assign(source='CellMarker')

mdf = (pd.concat([long1, long2], ignore_index=True)
       .drop_duplicates()
       .sort_values(['cell_type', 'gene'])
       .reset_index(drop=True))

print(mdf.shape)
mdf.head()
```

```python
# add categories
ct = (mdf["cell_type"]
      .astype("string")
      .str.normalize("NFKC")
      .str.strip()
      .str.replace(r"\s+", " ", regex=True)
      .str.lower()
)

mdf["category"] = np.select(
    [
        ct.str.contains(r"\bmyofibroblasts?\b", regex=True, na=False),
        ct.str.contains(r"\bfibroblasts?\b", regex=True, na=False),
        ct.str.contains(r"\b(myocytes?|muscles?|myoblasts?)\b", regex=True, na=False),
    ],
    ["myofibroblast", "fibroblast", "muscle"],
    default=pd.NA
)

mdf.head()
```

```python
gene_cats = (mdf.dropna(subset=["category"])
             .groupby("gene")["category"]
             .agg(lambda s: sorted(set(s))))

# 2) attach and compute "other_categories?"
tmp = mdf.join(gene_cats.rename("all_categories"), on="gene")

def other_cats(row):
    cats = row["all_categories"]
    cat = row["category"]
    if not isinstance(cats, list) or pd.isna(cat):
        return pd.NA
    if len(cats) == 1:
        return "unique"
    others = [c for c in cats if c != cat]
    return ", ".join(others) if others else "unique"

mdf["other_categories?"] = tmp.apply(other_cats, axis=1)
mdf.head()
```
