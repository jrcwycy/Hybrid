## MYOD1 subgraphs

```python
# ---------- Filter for MYOD nodes ----------
mdf = df[df['gene_name'].isin(GM.nodes())]
print(f"Shape after filtering for MYOD1 nodes: {mdf.shape=}")
# mdf.head()

# ---------- Add % expressing per condition columns ----------
def pct_expressing_by_group(adata, group_key, genes, layer="raw_counts", threshold=0):
    # subset to genes of interest
    genes = [g for g in genes if g in adata.var_names]
    A = adata[:, genes].layers[layer]
    if not sp.issparse(A):
        A = sp.csr_matrix(A)

    groups = adata.obs[group_key].astype(str).values
    out = pd.DataFrame(index=genes)

    for g in np.unique(groups):
        idx = np.where(groups == g)[0]
        sub = A[idx, :]
        pct = (sub > threshold).mean(axis=0)  # fraction expressing
        pct = np.asarray(pct).ravel()
        out[f"pct_expr_{g}"] = pct

    out.index.name = "gene_name"
    return out.reset_index()

# compute for MYOD network genes
pct_df = pct_expressing_by_group(
    adata, group_key="pooled_condition", genes=list(GM.nodes()), layer="raw_counts", threshold=0
)

mdf = mdf.merge(pct_df, on='gene_name', how='left')
print(mdf.shape)
mdf.head()
```

### Expression bins

```python
# ---------- Bin expression values ----------
# Quantile-based thresholding for TFs and non-TFs separately.
# Thresholds determined from Control cells (only expressed genes).

expr_cols = ['Control', 'mmMYOD1', 'siPRRX1', 'siPRRX1/mmMYOD1']
ref_col = 'Control'
tf_col = 'is_tf'
q_low = 1/3
q_high = 2/3 # tertiles -> low/med/high


# Control-only cutoffs, separately for TF vs non-TF
cutoffs = {}
for tf_flag, label in [(True, "tf"), (False, "non_tf")]:
    ref = mdf.loc[mdf[tf_col].eq(tf_flag), ref_col].astype(float)

    # learn cutoffs among expressed genes only
    ref_nz = ref[ref > 0]
    lo = ref_nz.quantile(q_low) if len(ref_nz) else 0.0
    hi = ref_nz.quantile(q_high) if len(ref_nz) else 0.0
    cutoffs[label] = (lo, hi)

def expr_bin_notexpr(x, lo, hi):
    x = x.astype(float)
    out = pd.Series(index=x.index, dtype="object")
    out.loc[x == 0] = "NE"
    out.loc[x > 0] = pd.cut(
        x.loc[x > 0],
        bins=[0, lo, hi, np.inf],
        labels=["low", "med", "high"],
        include_lowest=True
    ).astype(str)
    return out

# assign bins per condition using the Control-derived TF/non-TF cutoffs
for c in expr_cols:
    lo, hi = cutoffs["tf"]
    mdf.loc[mdf[tf_col].eq(True), f"expr_bin_{c}"] = expr_bin_notexpr(
        mdf.loc[mdf[tf_col].eq(True), c], lo, hi
    )

    lo, hi = cutoffs["non_tf"]
    mdf.loc[mdf[tf_col].eq(False), f"expr_bin_{c}"] = expr_bin_notexpr(
        mdf.loc[mdf[tf_col].eq(False), c], lo, hi
    )
    
print(mdf.shape)
mdf.head()
```

### Normalize expression

```python
# ---------- Normalize expression ----------

def robust_minmax_from_control_nonzero(mdf, expr_col, ref_col="Control", is_tf_col="is_tf",
                                       q_low=0.01, q_high=0.99, eps=1e-12):
    scaled = np.zeros(len(mdf), dtype=float)

    for tf_flag in [True, False]:
        mask = mdf[is_tf_col].eq(tf_flag).to_numpy()
        ref = mdf.loc[mask, ref_col].astype(float)

        ref_nz = ref[ref > 0]
        if len(ref_nz) == 0:
            # everything is zero in control for this group
            scaled[mask] = 0.0
            continue

        lo = ref_nz.quantile(q_low)
        hi = ref_nz.quantile(q_high)
        print(lo, hi)
        denom = (hi - lo) if hi > lo else 1.0

        x = mdf.loc[mask, expr_col].astype(float).to_numpy()

        s = (x - lo) / denom
        s[x <= 0] = 0.0                 # keep true zeros at 0
        scaled[mask] = np.clip(s, 0, 1)

    return scaled

cols_to_keep = ['gene_name'] + expr_cols + ['is_tf']
nmdf = mdf[cols_to_keep].copy()

nmdf["a_Control"] = robust_minmax_from_control_nonzero(nmdf, "Control")
nmdf["a_mmMYOD1"] = robust_minmax_from_control_nonzero(nmdf, "mmMYOD1")
nmdf["a_siPRRX1"] = robust_minmax_from_control_nonzero(nmdf, "siPRRX1")
nmdf["a_HYB"] = robust_minmax_from_control_nonzero(nmdf, "siPRRX1/mmMYOD1")

nmdf.head()
```

### Create weighted subgraphs

```python
# create subgraph with only expressed nodes per condition, then assign weights
def add_edge_weights_from_node_activity(G, a, weight_attr="weight", combine="product"):
    Gw = G.copy()
    for u, v in Gw.edges():
        au = a.get(u, 0.0)
        av = a.get(v, 0.0)
        if combine == "product":
            w = au * av
        elif combine == "min":
            w = min(au, av)
        elif combine == "mean":
            w = 0.5 * (au + av)
        else:
            raise ValueError("combine must be 'product', 'min', or 'mean'")
        Gw[u][v][weight_attr] = float(w)
    return Gw

def edge_weight_series(G):
    return [d.get("weight", 0.0) for _, _, d in G.edges(data=True)]


# ---------- Create weighted subgraphs ----------
# Control
ctrl_nodes = nmdf[nmdf['Control'] != 0]['gene_name'].unique()
CGM = GM.subgraph(ctrl_nodes).copy()
n_edges = CGM.number_of_edges()
n_nodes = CGM.number_of_nodes()
print(f"Control - N nodes: {n_nodes}, N edges: {n_edges}")

a_ctrl = dict(zip(nmdf["gene_name"], nmdf["a_Control"]))
wCGM = add_edge_weights_from_node_activity(CGM, a_ctrl, weight_attr="weight", combine="product")


# mmMYOD1
mmy_nodes = nmdf[nmdf['mmMYOD1'] != 0]['gene_name'].unique()
MGM = GM.subgraph(mmy_nodes).copy()
n_edges = MGM.number_of_edges()
n_nodes = MGM.number_of_nodes()
print(f"mmMYOD1 - N nodes: {n_nodes}, N edges: {n_edges}")

a_mmy = dict(zip(nmdf["gene_name"], nmdf["a_mmMYOD1"]))
wMGM = add_edge_weights_from_node_activity(MGM, a_mmy, weight_attr="weight", combine="product")


# siPRRX1
prrx_nodes = nmdf[nmdf['siPRRX1'] != 0]['gene_name'].unique()
PGM = GM.subgraph(prrx_nodes).copy()
n_edges = PGM.number_of_edges()
n_nodes = PGM.number_of_nodes()
print(f"siPRRX1 - N nodes: {n_nodes}, N edges: {n_edges}")

a_prrx = dict(zip(nmdf["gene_name"], nmdf["a_siPRRX1"]))
wPGM = add_edge_weights_from_node_activity(PGM, a_prrx, weight_attr="weight", combine="product")


# siPRRX1/mmMYOD1
hyb_nodes = nmdf[nmdf['siPRRX1/mmMYOD1'] != 0]['gene_name'].unique()
HGM = GM.subgraph(hyb_nodes).copy()
n_edges = HGM.number_of_edges()
n_nodes = HGM.number_of_nodes()
print(f"siPRRX1/mmMYOD1 - N nodes: {n_nodes}, N edges: {n_edges}")

a_hyb = dict(zip(nmdf["gene_name"], nmdf["a_HYB"]))
wHGM = add_edge_weights_from_node_activity(HGM, a_hyb, weight_attr="weight", combine="product")


# ---------- Sum edge weights ----------


ctrl_sum = sum(edge_weight_series(wCGM))
print(f"\nSum edge weight (Control): {ctrl_sum:.3f}")

m_sum = sum(edge_weight_series(wMGM))
print(f"Sum edge weight (mmMYOD1): {m_sum:.3f}")

p_sum = sum(edge_weight_series(wPGM))
print(f"Sum edge weight (siPRRX1): {p_sum:.3f}")

h_sum = sum(edge_weight_series(wHGM))
print(f"Sum edge weight (siPRRX1/mmMYOD1): {h_sum:.3f}")
```

```python
# ---------- Bin expression values ----------
# Quantile-based thresholding for TFs and non-TFs separately.
# Thresholds determined from Control cells (only expressed genes).

expr_cols = ['Control', 'mmMYOD1', 'siPRRX1', 'siPRRX1/mmMYOD1']
ref_col = 'Control'
tf_col = 'is_tf'
q_low = 1/3
q_high = 2/3 # tertiles -> low/med/high


# Control-only cutoffs, separately for TF vs non-TF
cutoffs = {}
for tf_flag, label in [(True, "tf"), (False, "non_tf")]:
    ref = pdf.loc[pdf[tf_col].eq(tf_flag), ref_col].astype(float)

    # learn cutoffs among expressed genes only
    ref_nz = ref[ref > 0]
    lo = ref_nz.quantile(q_low) if len(ref_nz) else 0.0
    hi = ref_nz.quantile(q_high) if len(ref_nz) else 0.0
    cutoffs[label] = (lo, hi)

def expr_bin_notexpr(x, lo, hi):
    x = x.astype(float)
    out = pd.Series(index=x.index, dtype="object")
    out.loc[x == 0] = "NE"
    out.loc[x > 0] = pd.cut(
        x.loc[x > 0],
        bins=[0, lo, hi, np.inf],
        labels=["low", "med", "high"],
        include_lowest=True
    ).astype(str)
    return out

# assign bins per condition using the Control-derived TF/non-TF cutoffs
for c in expr_cols:
    lo, hi = cutoffs["tf"]
    pdf.loc[pdf[tf_col].eq(True), f"expr_bin_{c}"] = expr_bin_notexpr(
        pdf.loc[pdf[tf_col].eq(True), c], lo, hi
    )

    lo, hi = cutoffs["non_tf"]
    pdf.loc[pdf[tf_col].eq(False), f"expr_bin_{c}"] = expr_bin_notexpr(
        pdf.loc[pdf[tf_col].eq(False), c], lo, hi
    )
    
print(pdf.shape)
pdf.head()
```

### Normalize expression

```python
# ---------- Normalize expression ----------

cols_to_keep = ['gene_name'] + expr_cols + ['is_tf']
npdf = pdf[cols_to_keep].copy()

npdf["a_Control"] = robust_minmax_from_control_nonzero(npdf, "Control")
npdf["a_mmMYOD1"] = robust_minmax_from_control_nonzero(npdf, "mmMYOD1")
npdf["a_siPRRX1"] = robust_minmax_from_control_nonzero(npdf, "siPRRX1")
npdf["a_HYB"] = robust_minmax_from_control_nonzero(npdf, "siPRRX1/mmMYOD1")

npdf.head()
```

```python
# ---------- Create weighted subgraphs ----------
# Control
ctrl_nodes = npdf[npdf['Control'] != 0]['gene_name'].unique()
CGP = GP.subgraph(ctrl_nodes).copy()
n_edges = CGP.number_of_edges()
n_nodes = CGP.number_of_nodes()
print(f"Control - N nodes: {n_nodes}, N edges: {n_edges}")

a_ctrl = dict(zip(npdf["gene_name"], npdf["a_Control"]))
wCGP = add_edge_weights_from_node_activity(CGP, a_ctrl, weight_attr="weight", combine="product")


# mmMYOD1
mmy_nodes = npdf[npdf['mmMYOD1'] != 0]['gene_name'].unique()
MGP = GP.subgraph(mmy_nodes).copy()
n_edges = MGP.number_of_edges()
n_nodes = MGP.number_of_nodes()
print(f"mmMYOD1 - N nodes: {n_nodes}, N edges: {n_edges}")

a_mmy = dict(zip(npdf["gene_name"], npdf["a_mmMYOD1"]))
wMGP = add_edge_weights_from_node_activity(MGP, a_mmy, weight_attr="weight", combine="product")


# siPRRX1
prrx_nodes = npdf[npdf['siPRRX1'] != 0]['gene_name'].unique()
PGP = GP.subgraph(prrx_nodes).copy()
n_edges = PGP.number_of_edges()
n_nodes = PGP.number_of_nodes()
print(f"siPRRX1 - N nodes: {n_nodes}, N edges: {n_edges}")

a_prrx = dict(zip(npdf["gene_name"], npdf["a_siPRRX1"]))
wPGP = add_edge_weights_from_node_activity(PGP, a_prrx, weight_attr="weight", combine="product")


# siPRRX1/mmMYOD1
hyb_nodes = npdf[npdf['siPRRX1/mmMYOD1'] != 0]['gene_name'].unique()
HGP = GP.subgraph(hyb_nodes).copy()
n_edges = HGP.number_of_edges()
n_nodes = HGP.number_of_nodes()
print(f"siPRRX1/mmMYOD1 - N nodes: {n_nodes}, N edges: {n_edges}")

a_hyb = dict(zip(npdf["gene_name"], npdf["a_HYB"]))
wHGP = add_edge_weights_from_node_activity(HGP, a_hyb, weight_attr="weight", combine="product")


# ---------- Sum edge weights ----------


ctrl_sum = sum(edge_weight_series(wCGP))
print(f"\nSum edge weight (Control): {ctrl_sum:.3f}")

m_sum = sum(edge_weight_series(wMGP))
print(f"Sum edge weight (mmMYOD1): {m_sum:.3f}")

p_sum = sum(edge_weight_series(wPGP))
print(f"Sum edge weight (siPRRX1): {p_sum:.3f}")

h_sum = sum(edge_weight_series(wHGP))
print(f"Sum edge weight (siPRRX1/mmMYOD1): {h_sum:.3f}")
```

The independence index statistic $ρ^k_{ij}$

Used for creating single-cell gene correlation networks

$ρ^k_{ij} \geq 0.01$ means an edge exists between genes i and j in cell k

Defined as the difference between the observed joint probability and the expected joint probability under independence (analogous to a covariance but estimated non-parametrically using neighborhood boxes)

```python
gene_i = 'ACTB'
gene_j = 'GAPDH'
tmp = X_df[[gene_i, gene_j]]

print(f"Total cells: {tmp.shape[0]}")

# filter for cells expressing both genes
tmp = tmp[(tmp[gene_i] > 0) & (tmp[gene_j] > 0)]
print(f"Total cells with non-zero expression for both genes: {tmp.shape[0]}")

total_cells = tmp.shape[0]
neighborhood_size = int(0.1 * total_cells)
print(f"Neighborhood size = {neighborhood_size}")

x = tmp[gene_i].values  # numpy arrays for speed
y = tmp[gene_j].values
cell_ids = tmp.index


def compute_p_for_cell(k):
    xk = x[k]
    yk = y[k]

    # X neighborhood bounds
    x_diffs = np.abs(x - xk)
    x_nearest_idx = np.argpartition(x_diffs, neighborhood_size)[:neighborhood_size]
    x_min, x_max = x[x_nearest_idx].min(), x[x_nearest_idx].max()

    # Y neighborhood bounds
    y_diffs = np.abs(y - yk)
    y_nearest_idx = np.argpartition(y_diffs, neighborhood_size)[:neighborhood_size]
    y_min, y_max = y[y_nearest_idx].min(), y[y_nearest_idx].max()

    # Box masks
    x_box_mask = (x >= x_min) & (x <= x_max)
    y_box_mask = (y >= y_min) & (y <= y_max)
    intersection_mask = x_box_mask & y_box_mask

    n_x_box = x_box_mask.sum()
    n_y_box = y_box_mask.sum()
    n_intersection = intersection_mask.sum()

    p = (n_intersection / total_cells) - (n_x_box / total_cells) * (n_y_box / total_cells)
    return p, x_min, x_max, y_min, y_max

# Run for all cells
results = [compute_p_for_cell(k) for k in range(total_cells)]

p_values, x_mins, x_maxs, y_mins, y_maxs = zip(*results)

result_df = pd.DataFrame({
    'cell_id': cell_ids,
    'p': p_values
})

bounds_df = pd.DataFrame({
    'cell_id': cell_ids,
    'x_min': x_mins,
    'x_max': x_maxs,
    'y_min': y_mins,
    'y_max': y_maxs,
})

print(result_df.shape)
display(result_df.head())
display(bounds_df.head())
```

```python
def plot_scatters(cells_to_plot, title):
    
    ncols = 5
    nrows = int(len(cells_to_plot) / ncols)
    
    fig, axs = plt.subplots(nrows, ncols, figsize=(ncols*3,nrows*3.5))
    axs = axs.flatten()
    
    for i, cell in enumerate(cells_to_plot):
        ax = axs[i]

        ax.scatter(x, y, s=1)

        k = tmp.index.get_loc(cell)
        xk, yk = x[k], y[k]
        ax.scatter(xk, yk, s=10, color='red', edgecolor='k', zorder=1)

        bounds = bounds_df.loc[bounds_df['cell_id'] == cell].iloc[0]
        x_min, x_max = bounds['x_min'], bounds['x_max']
        y_min, y_max = bounds['y_min'], bounds['y_max']   

        rect_x = Rectangle(
            (x_min, y.min()),
            x_max - x_min,
            y.max() - y.min(),
            facecolor='blue', alpha=0.2, edgecolor='none', zorder=10,
        )
        rect_y = Rectangle(
            (x.min(), y_min),
            x.max() - x.min(),
            y_max - y_min,
            facecolor='green', alpha=0.2, edgecolor='none', zorder=10,
        )
        ax.add_patch(rect_x)
        ax.add_patch(rect_y)

        ax.set_xlabel(f"{gene_i}")
        ax.set_ylabel(f"{gene_j}")
        ax.set_title(f"k = {cell}", fontsize=9)
        
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
    
plot_scatters(top_cells, title='$ρ^k_{ij}$ > 0.01 — positive co-expression (5 highest $ρ_{ij}$)')
plot_scatters(bottom_cells, title='$ρ^k_{ij}$ > 0.01 — positive co-expression (5 lowest $ρ_{ij}$)')
plot_scatters(nonsig_cells, title='$ρ^k_{ij}$ < 0 — negative association (5 lowest $ρ_{ij}$)')
plot_scatters(nearzero_cells, title='$ρ^k_{ij}$ ≈ 0 — independent expression (5 $ρ_{ij}$ closest to 0)')
```
